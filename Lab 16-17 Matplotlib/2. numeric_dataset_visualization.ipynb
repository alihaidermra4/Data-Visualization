## Visualizing Numeric Data

In this notebok, we will visualize a dataset consists of `numeric data`.

### Iris Dataset
The Iris dataset is a well-known dataset used for machine learning and data visualization tasks. It consists of `150 observations of three iris species:` Setosa, Versicolor, and Virginica.

The dataset includes four features: sepal length, sepal width, petal length, and petal width, measured in centimeters. These features are commonly used to classify the species of each iris flower.

![iris_dataset.png](iris_dataset.png)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris
```

**Bar Chart (Average Petal Length by Species)**

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='petal_length', data=iris, palette='Set2')
plt.title("Average Petal Length by Species")
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

**Line Plot (Sepal length across samples)**

```python
plt.figure(figsize=(8, 5))
plt.plot(iris.index, iris['sepal_length'], marker='o', color='blue', linestyle='-')
plt.title('Sepal Length Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
iris.index
```

**Scatter Plot (Sepal Length vs Petal Length)**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris, palette='deep')
plt.title('Sepal Length vs Petal Length')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

**Histogram (Distribution of Sepal Width)**

```python
plt.figure(figsize=(8, 5))
plt.hist(iris['sepal_width'], bins=10, color='lightgreen', alpha=0.8)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

---

**`hue`**  In Seaborn, `hue` is used to group the data and apply different colors to the distict groups.

---

**`palette`** refers to the color scheme used in seaborn to color the data. It allows you to specify a set of colors to be used in plots. For example:
* `palette='Blues':` A gradient of blue shades.
* `palette='coolwarm':` A diverging color scheme from cool (blue) to warm (red).
* `palette='Set2':` A predefined set of colors that works well for categorical data.
